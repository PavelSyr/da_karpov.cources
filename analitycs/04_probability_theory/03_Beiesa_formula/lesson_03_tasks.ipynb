```python
from scipy import special #модуль питона, предназначенный для научных и инженерных расчетов
# special.comb(n, k) #биномиальный коэффициент

import numpy as np
```

```python
def full_pobability(ns, ps):
    n = np.sum(ns)
    
    p = 0

    for i in range(0, len(ns)):
        ni = ns[i]
        pi = ps[i]
        p += ni/n * pi
    
    return p

def beiesa(p_a, n, p, m):
    return (n/m * p) / p_a

def bernuli(n, k, p, q):
    c = special.comb(n, k)
    return c * (p**k) * (q**(n-k))
```

В любой момент времени вероятность того, что самолет находится в зоне действия радара, составляет 7%.Какова вероятность того, что самолет находится в зоне действия радара, учитывая, что на радаре нет самолета?
